```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
file = open('partA_2/time_analysis_average.txt','r')
content = [data.split(',') for data in json.load(file)]
df = pd.DataFrame(content, columns = ['Date', 'Average'])
df['Average'] = df['Average'].astype(float) 
df = df.sort_values(by=['Average'], ascending=False)
df.head()
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation = 90)
plt.xlabel('Month / Year')
plt.ylabel('Average')
plt.bar(df['Date'],  df['Average'])
```
