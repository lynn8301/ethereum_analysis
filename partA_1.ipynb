```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
file = open('partA_1/time_analysis_total.txt','r')
content = json.load(file)
df = pd.DataFrame(content, columns = ['Date', 'Transactions'])
df = df.sort_values(by = ['Transactions'], ascending=False)
df.head()
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation = 90)
plt.xlabel('Date')
plt.ylabel('Total Transactions')
plt.bar(df['Date'],  df['Transactions'])
```
