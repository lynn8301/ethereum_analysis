```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
file = open('partD_Gas_Guzzlers/average_gas_price.txt','r')
content = json.load(file)
df = pd.DataFrame(content, columns = ['Date', 'Average Gas Price'])
df['Average Gas Price'] = df['Average Gas Price'].astype(float) 
df['Date'] = pd.to_datetime(df['Date'], format='%m/%Y')
df = df.sort_values(by='Date')
df.head()
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation = 0)
plt.xlabel('Date')
plt.ylabel('Average Gas Price')
plt.title('Average Gas Price')
plt.plot(df['Date'],  df['Average Gas Price'], color='green')
```

```python
file = open('partD_Gas_Guzzlers/average_gas_used.txt','r')  
content = json.load(file)
df = pd.DataFrame(content, columns = ['Date', 'Average Gas Used'])
df['Average Gas Used'] = df['Average Gas Used'].astype(float) 
df['Date'] = pd.to_datetime(df['Date'], format='%m/%Y')
df = df.sort_values(by='Date')
df.head()
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation = 90)
plt.xlabel('Date')
plt.ylabel('Average Gas Used')
plt.title('Average Gas Used')
plt.plot(df['Date'],  df['Average Gas Used'], color='orange')
```
